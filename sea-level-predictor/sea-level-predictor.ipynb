```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
fig,ax = plt.subplots(figsize=(15,8))
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])

res = linregress(df['Year'],df['CSIRO Adjusted Sea Level'])
prediction = pd.Series([i for i in range(1880,2050)])
plt.plot(prediction,res.intercept + res.slope*prediction,'r')

df_new_line = df.loc[df['Year'] >= 2000]
res_new_line = linregress(df_new_line['Year'],df_new_line['CSIRO Adjusted Sea Level'])
prediction_new_line = pd.Series([i for i in range(2000,2050)])
plt.plot(prediction_new_line,res_new_line.intercept + res_new_line.slope*prediction_new_line,'orange')

ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
```
